```python
from allennlp.commands.elmo import ElmoEmbedder
from pathlib import Path
import torch

# Load the pre-trained SeqVec model
model_dir = Path('path_to_pretrained_SeqVec_directory')
weights = model_dir / 'weights.hdf5'
options = model_dir / 'options.json'
seqvec = ElmoEmbedder(options, weights, cuda_device=0)  # Use cuda_device=-1 for CPU

# Function to get embedding for a single amino acid sequence
def get_embedding(sequence):
    embedding = seqvec.embed_sentence(list(sequence))  # List-of-Lists with shape [3, L, 1024]
    return torch.tensor(embedding).sum(dim=0).mean(dim=0)  # Vector with shape [1024]

# Example usage
sequence = 'ACDEFGHIKLMNPQRSTVWY'
embedding = get_embedding(sequence)
print(embedding)
```

## ESM :

### Meth 1 :

```python
!pip install fair-esm
```

```python
import torch
import esm

# Load the ESM model and alphabet
model, alphabet = esm.pretrained.esm1b_t33_650M_UR50S()
model.eval()

# Define teh amino acid sequence
sequence = "MENKAPKLCKALSKYGGGSGGGSGGSGGSGGLFET"
sequence = sequence.upper()

# Tokenize the sequence
batch_converter = alphabet.get_batch_converter()
data = [("protein1", sequence)]
batch_labels, batch_strs, batch_tokens = batch_converter(data)

# Forward pass to get embeddings
with torch.no_grad():
    results = model(batch_tokens, repr_layers=[33])
    token_representations = results["representations"][33]

# Get the per-residue embeddings (excluding the padding, start, and end tokens)
sequence_embeddings = token_representations[0, 1:-1].mean(0)

print("Sequence embedding shape:", sequence_embeddings.shape)
print("Sequence embedding:", sequence_embeddings)
```

```python
from transformers import EsmTokenizer, EsmModel
import torch

# Load the tokenizer and the model
tokenizer = EsmTokenizer.from_pretrained("facebook/esm1b-t33_650M_UR50S")
model = EsmModel.from_pretrained("facebook/esm1b-t33_650M_UR50S")

# Example amino acid sequence
sequence = "MKTAYIAKQRQISFVKSHFSRQLEERLGLIEVQIDNEDLFRHMVYARDSTLRTKNNYTEEKNRLN"

# Tokenize the sequence
inputs = tokenizer(sequence, return_tensors="pt", add_special_tokens=False)

# Get the embeddings
with torch.no_grad():
    outputs = model(**inputs)

# Extract the embeddings from the model output
embeddings = outputs.last_hidden_state

# Print the shape of the embeddings
print(embeddings.shape)
```

```python
# Load the tokenizer and model
model_name = "facebook/esm2_t33_650M_UR50D"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = EsmModel.from_pretrained(model_name)
```
